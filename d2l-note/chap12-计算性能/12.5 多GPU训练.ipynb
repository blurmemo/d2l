```python
import torch
from torch import nn
from torch.nn import functional as F
from d2l import torch as d2l
```

```python
scale = 0.01
# 卷积核、全连接层权重和偏差参数初始化
W1 = torch.rand(size=(20, 1, 3, 3)) * scale
b1 = torch.zeros(20)
W2 = torch.rand(size=(50, 20, 5, 5)) * scale
b2 = torch.zeros(50)
W3 = torch.rand(size=(800, 128)) * scale
b3 = torch.zeros(128)
W4 = torch.rand(size=(128, 10)) * scale
b4 = torch.zeros(10)
params = [W1, b1, W2, b2, W3, b3, W4, b4]
```

```python
# 定义模型
def lenet(X, params):
    h1_conv = F.conv2d(X, weight=params[0], bias=params[1])
    h1_activation = F.relu(h1_conv)
    h1 = F.avg_pool2d(input=h1_activation, kernel_size=(2, 2), stride=(2, 2))
    h2_conv = F.conv2d(h1, weight=params[2], bias=params[3])
    h2_activation = F.relu(h2_conv)
    h2 = F.avg_pool2d(input=h2_activation, kernel_size=(2, 2), stride=(2, 2))
    h2 = h2.reshape(h2.shape[0], -1)
    h3_linear = torch.mm(h2, params[4]) + params[5]
    h3 = F.relu(h3_linear)
    y_hat = torch.mm(h3, params[6]) + params[7]
    return y_hat
```

```python
loss = nn.CrossEntropyLoss(reduction='none')
```

```python
# 分发模型参数到GPU i
def get_params(params, device):
    new_params = [p.to(device) for p in params]
    for p in new_params:
        p.requires_grad_()
    return new_params
```

```python
new_params = get_params(params, d2l.try_gpu(0))
print('b1 权重', new_params[1])
print('b1 梯度', new_params[1].grad)
```

```python
def allreduce(data):
    for i in range(1, len(data)):
        data[0][:] += data[i].to(data[0].device)
    for i in range(1, len(data)):
        data[i][:] = data[0].to(data[i].device)
```

```python
data = [torch.ones((1, 2), device=d2l.try_gpu(i)) * (i + 1) for i in range(2)]
print('allreduce之前: \n', data[0], '\n', data[1])
allreduce(data)
print('allreduce之后: \n', data[0], '\n', data[1])
```

```python
# 数据分发
data = torch.arange(20).reshape(4, 5)
devices = [torch.device('cuda:0'), torch.device('cuda:1')]
split = nn.parallel.scatter(data, devices)
print('input: ', data)
print('load into', devices)
print('output: ', split)
```

```python
def split_batch(X, y, devices):
    assert X.shape[0] == y.shape[0]
    return (
        nn.parallel.scatter(X, devices),
        nn.parallel.scatter(y, devices)
    )
```

```python
def train_batch(X, y, device_params, devices, lr):
    X_shards, y_shards = split_batch(X, y, devices)
    # 在每一个GPU上计算loss
    ls = [loss(lenet(X_shard, device_param), y_shard).sum() for X_shard, y_shard, device_param in zip(X_shards, y_shards, device_params)]
    for l in ls:
        l.backward()
    with torch.no_grad():
        # W，b参数遍历
        for i in range(len(device_params[0])):
            # 对每个GPU进行梯度相加
            allreduce([device_params[c][i].grad for c in range(len(devices))])
        for param in device_params:
            d2l.sgd(param, lr, X.shape[0])
```

```python
def train(num_gpus, batch_size, lr):
    train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
    devices = [d2l.try_gpu(i) for i in range(num_gpus)]
    device_params = [get_params(params ,d) for d in devices]
    num_epochs = 10
    animator = d2l.Animator('epoch', 'test_acc', xlim=[1, num_epochs])
    timer = d2l.Timer()
    for epoch in range(num_epochs):
        timer.start()
        for X, y in train_iter:
            train_batch(X, y, device_params, devices, lr)
            torch.cuda.synchronize()
        timer.stop()
        # 在GPU 0 上评估
        animator.add(epoch + 1, (d2l.evaluate_accuracy_gpu(lambda x: lenet(x, device_params[0]), test_iter, devices[0]),))
    print(f'测试精度：{animator.Y[0][-1]:.2f}, {timer.avg():.1f}秒/轮, 在{str(devices)}')
```

```python
train(num_gpus=1, batch_size=256, lr=0.2)
```
